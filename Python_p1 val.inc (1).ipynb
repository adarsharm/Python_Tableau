# Addition of CostPerTransaction and SalesPerTransaction column


```python
data["CostPerTransaction"]=data["CostPerItem"]*data["NumberOfItemsPurchased"]
data["SalesPerTransaction"]=data["SellingPricePerItem"]*data["NumberOfItemsPurchased"]
```

```python
data.info()
```

# Addition of ProfitPerTransaction and Markup Column

```python
data["ProfitPerTransaction"]=data["SalesPerTransaction"]-data["CostPerTransaction"]
data["Markup"]=(data["ProfitPerTransaction"])/data["CostPerTransaction"]
#markup=(salespertransation-costpertransaction)/costpertransaction
```

```python
data.info()
```

```python
#round off markup to 2
data["Markup"]=round(data["Markup"],2)
```

# clubbing date to one column 

```python
#day and year column is int type but '-' and month is string type so we will have to change the datatype of day and year
day= data["Day"].astype(str)
print(day.dtype)
```

```python
year=data["Year"].astype(str)
date= day +"-"+data["Month"]+"-"+year
data["date"]=date
```

```python
# using split to split the client keywords field
#new_var=column.str.split("sep",expand=True)
split_col=data["ClientKeywords"].str.split(",",expand=True)
data["ClientAge"]=split_col[0]
data["ClientType"]=split_col[1]
data["LengthOfContract"]=split_col[2]
                             
```

```python
data["LengethOfContract"]=data["LengthOfContract"].str.replace("]","")
data["ClientAge"]=data["ClientAge"].str.replace("[","")
```

```python
#using lower function to change item description to lower case
data["ItemDescription"]=data["ItemDescription"].str.lower()
```
